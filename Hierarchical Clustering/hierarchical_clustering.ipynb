# Hierarchical Clustering 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset 

```python
df = pd.read_csv('Mall_Customers.csv')
X = df.iloc[:,[3,4]].values
print(X)
```

## Using Dendogram to find optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
```

```python
dendogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendrogram')
```

## Training model 

```python
from sklearn.cluster import  AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='ward')
```

```python
y_hc = hc.fit_predict(X)
```

```python
print(y_hc)
```

```python
plt.scatter(X[y_hc ==0,0],X[y_hc ==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X[y_hc ==1,0],X[y_hc ==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(X[y_hc ==2,0],X[y_hc ==2,1],s=100,c='purple',label='Cluster 3')
plt.scatter(X[y_hc ==3,0],X[y_hc ==3,1],s=100,c='green',label='Cluster 4')
plt.scatter(X[y_hc ==4,0],X[y_hc ==4,1],s=100,c='cyan',label='Cluster 5')
plt.title('Clusters of Customer')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Scores (1-100)')
plt.legend()
plt.show()
```
